```python
cities = pd.read_csv("TERC_Urzedowy_2020-01-28.csv", header=0, sep=";")
cities.fillna(0,inplace=True)
cities.NAZWA = cities.NAZWA.str.lower()
dictionary = {}

for i in range(len(cities)):
    if cities.POW[i] == 0:
        dictionary[cities.WOJ[i]] = cities.NAZWA[i] 
cities_c = {}

for i in range(len(cities)):
    cities_c[cities.NAZWA[i]] = dictionary[cities.WOJ[i]]
```

```python
for i in range(len(df_k)):
    try:
        df_k.iloc[i:i+1,11:12] = cities_c[df_k['Nazwa stacji'][i]]  #przypisanie do wojewodztw --> 
    except:
        df_k.iloc[i:i+1,11:12] = 0
        
df_k.to_csv("Klimat_woj.csv", index=False)
df_k.head()
```
